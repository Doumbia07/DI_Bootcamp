# Loan Default Prediction: Problem Definition and Data Collection Plan

The objective of this project is to predict whether a loan applicant is likely to default on a loan. The target variable is loan default status (default or non-default), and the prediction will support lending decisions by helping financial institutions assess risk before approving loans.

Several types of data are required for this project. Personal information such as age, marital status, education level, and employment status can help describe the applicant’s socioeconomic profile. Financial information such as annual income, monthly income, debt-to-income ratio, and savings can help assess the applicant’s ability to repay the loan. Credit-related data such as credit score, credit utilization, number of open accounts, and previous delinquencies are important because they reflect past financial behavior and creditworthiness. Loan-specific information such as loan amount, interest rate, loan term, and loan purpose is also necessary because these factors directly influence repayment obligations. Repayment history data, including previous loan repayments, late payments, and defaults, can provide strong indicators of future repayment behavior.

The primary source of data would be the financial institution’s internal records, which contain loan applications, repayment histories, customer profiles, and transaction data. Additional credit-related information can be obtained from credit bureaus, which provide credit scores, delinquency records, and overall credit histories. Customer-provided application forms can supply demographic, employment, and income information. External economic and demographic data may be collected from government agencies or public statistical databases to provide contextual information about employment rates, income levels, and regional economic conditions. These different sources would be integrated using unique customer identifiers and standardized data formats.

Several risks and constraints must be considered during data collection and model development. Privacy regulations require that customer data be collected, stored, and processed securely while complying with applicable data protection laws. Data quality issues such as missing values, inaccurate records, and duplicate entries can negatively affect model performance and must be addressed through preprocessing. Sampling bias may occur if the dataset does not adequately represent all borrower groups, which can lead to unfair predictions. Strong data governance practices, including access controls, auditing procedures, and documentation standards, are necessary to ensure the reliability, transparency, and ethical use of the data and the resulting machine learning model.

```python
#exo3
# This piece of code is already prefilled, run it to execute it and see the results.
# It demonstrates standard classification metrics for binary loan default prediction.
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score, confusion_matrix, classification_report

# Please replace these placeholders with your true labels and predicted probabilities.
y_true = [0,1,0,1,0,0,1,0,1,0]            # placeholder labels
y_pred_proba = [0.05,0.80,0.10,0.65,0.20,0.15,0.70,0.30,0.85,0.25]  # placeholder probabilities

# You should set a decision threshold that reflects the precision–recall trade-off for your business case.
threshold = 0.5
y_pred = [1 if p >= threshold else 0 for p in y_pred_proba]

print("Accuracy:", round(accuracy_score(y_true, y_pred), 4))
print("Precision:", round(precision_score(y_true, y_pred, zero_division=0), 4))
print("Recall:", round(recall_score(y_true, y_pred, zero_division=0), 4))
print("F1-score:", round(f1_score(y_true, y_pred, zero_division=0), 4))
print("ROC-AUC:", round(roc_auc_score(y_true, y_pred_proba), 4))
print("PR-AUC (Average Precision):", round(average_precision_score(y_true, y_pred_proba), 4))
print("\nConfusion matrix:\n", confusion_matrix(y_true, y_pred))
print("\nClassification report:\n", classification_report(y_true, y_pred, zero_division=0))
```
